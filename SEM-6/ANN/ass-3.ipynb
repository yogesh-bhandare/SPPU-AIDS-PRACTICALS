```python
import numpy as np
```

```python
# Perceptron Learning Algorithm
def perceptron_learning(X, y, learning_rate=0.1, epochs=100):
    # Initialize weights and bias
    weights = np.zeros(X.shape[1])
    bias = 0
    
    for epoch in range(epochs):
        for i in range(X.shape[0]):
            # Compute the prediction
            prediction = np.dot(X[i], weights) + bias
            # Apply the step function (activation function)
            prediction = 1 if prediction >= 0 else -1
            
            # Update weights and bias if misclassified
            if prediction != y[i]:
                weights += learning_rate * y[i] * X[i]
                bias += learning_rate * y[i]
        
        # Print weights and bias for each epoch (optional)
        print(f"Epoch {epoch + 1}: Weights = {weights}, Bias = {bias}")
    
    return weights, bias
```

```python
# Function to classify a number as even or odd
def classify_number(number, weights, bias):
    # Convert the number to its ASCII value
    ascii_value = ord(str(number))
    # Compute the prediction
    prediction = np.dot(ascii_value, weights) + bias
    # Apply the step function
    return 1 if prediction >= 0 else -1
```

```python
# Example dataset (ASCII values for digits 0-9)
X = np.array([ord(str(i)) for i in range(10)]).reshape(-1, 1)  # ASCII values as features
y = np.array([1 if i % 2 == 0 else -1 for i in range(10)])     # Labels: 1 for even, -1 for odd

# Train the perceptron
weights, bias = perceptron_learning(X, y)
```

```python
# Print final weights and bias
print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Test the perceptron on numbers 0 to 9
print("\nTesting the Perceptron:")
for number in range(10):
    result = classify_number(number, weights, bias)
    print(f"Number: {number}, ASCII: {ord(str(number))}, Predicted: {'Even' if result == 1 else 'Odd'}")
```
